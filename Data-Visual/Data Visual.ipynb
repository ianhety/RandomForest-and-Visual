```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%pylab inline
sns.set_style('darkgrid')
```

```python
data = pd.read_csv("C:\\Users\\310249682\\Downloads\\WorldPhones.csv")
data.head()
```

```python
sns.set(font_scale=2.5) 
fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6))= plt.subplots(3, 2, sharex='col', sharey=False)
fig.set_size_inches(23, 20.5)
ax1.plot(data["Year"], data["N.Amer"], '-gD',markevery = [1,3])
ax1.set_ylabel('N.Amer')
ax2.plot(data["Year"], data["Europe"], '-gD',markevery = [0,2])
ax2.set_ylabel('Europe')
ax3.plot(data["Year"], data["Asia"], '-gD',markevery = [0,4])
ax3.set_ylabel('Asia')
ax4.plot(data["Year"], data["S.Amer"], '-gD',markevery = [1,2])
ax4.set_ylabel('S.Amer')
ax5.plot(data["Year"], data["Oceania"], '-gD',markevery = [1,3])
ax5.set_ylabel('Oceania')
ax5.set_xlabel('Year')
ax6.plot(data["Year"], data["Africa"], '-gD',markevery = [1,3])
ax6.set_ylabel('Africa')
ax6.set_xlabel('Year')
```
